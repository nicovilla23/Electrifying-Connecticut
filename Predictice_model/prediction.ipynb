```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
```

```python
df = pd.read_csv('Clean_Electric_Vehicle_Population_Data.csv')
```

```python

df = df[df['electric_range']!=0]
df = df[df['electric_vehicle_type']!='Plug-in Hybrid']
df= df[['model_year','electric_range']]
df.sort_values('model_year', ascending=True, inplace=True)
df = df[(df.model_year >= 2011) & (df.model_year<=2020)]
df.reset_index(inplace=True)
```

```python
df.drop(columns= 'index',inplace=True)
```

```python
df['model_year'] = df['model_year'].astype(int)


df.set_index('model_year', inplace=True)
```

```python
lop= pd.read_clipboard()
```

```python
pop= lop[['YEAR','BATTERY ELECTRIC CARS']]
pop
```

```python
pop.set_index('YEAR', inplace=True)
```

```python
pop
```

```python
autopop = auto_arima(pop.values,trace=True,suppress_warnings= True)
autopop.summary()
```

```python
modelpop = ARIMA(pop.values, order=(1, 0, 0))
model_fit_pop = modelpop.fit()


predictions = model_fit_pop.predict(start=2021, end=2023)


print(predictions)
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing



data = {'YEAR': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
        'pop': [591, 7179, 13730, 21454, 31197, 46857, 54065, 83491, 132377, 242966, 536186, 876527]}
df = pd.DataFrame(data)
df.set_index('YEAR', inplace=True)


model = ExponentialSmoothing(df['pop'], trend='add', seasonal='add', seasonal_periods=12)
model_fit = model.fit()
predictions = model_fit.predict(start=2021, end=2025)
print(predictions)
```

```python
lst = [e for e in range(2011,2021)]
avgs = []
yrs=[]
for e in lst:
    x=df[df['model_year']== e].mean().values[1]
    y =df[df['model_year']== e].mean().values[0]
    avgs.append(x)
    yrs.append(y)
```
