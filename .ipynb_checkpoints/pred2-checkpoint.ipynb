```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


# Assuming the data is in a variable called 'pop'
years = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021])
population = np.array([591, 7179, 13730, 21454, 31197, 46857, 54065, 83491, 132377, 242966, 536186, 876527])
pop_df = pd.DataFrame({'YEAR': years, 'Population': population})


plt.scatter(years, population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population over Time')
plt.show()

X = pop_df[['YEAR']]
y = pop_df['Population']

# Linear Regression Model
linear_reg = LinearRegression()
linear_reg.fit(X, y)

# Polynomial Regression Model (degree 2)
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
poly_reg_model = LinearRegression()
poly_reg_model.fit(X_poly, y)


# Create a range of years from 2021 to 2025
future_years = np.array([2021, 2022, 2023, 2024, 2025]).reshape(-1, 1)

# Make predictions using the linear regression model
linear_reg_predictions = linear_reg.predict(future_years)

# Make predictions using the polynomial regression model
future_years_poly = poly_reg.fit_transform(future_years)
poly_reg_predictions = poly_reg_model.predict(future_years_poly)


plt.scatter(years, population)
plt.plot(years, linear_reg.predict(X), label='Linear Regression')
plt.plot(years, poly_reg_model.predict(poly_reg.fit_transform(X)), label='Polynomial Regression')
plt.plot(future_years, linear_reg_predictions, 'ro', label='Linear Regression Predictions')
plt.plot(future_years, poly_reg_predictions, 'go', label='Polynomial Regression Predictions')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population over Time')
plt.legend()
plt.show()
print(poly_reg_predictions)
```
